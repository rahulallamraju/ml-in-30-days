```python
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
sns.scatterplot(df, x=df.sepal_width, y=df.petal_width)
```

```python
from sklearn.svm import OneClassSVM
ocsvm = OneClassSVM(nu=0.001)
X = df[['sepal_width', 'petal_width']]
ocsvm.fit(X)
ocsvm.predict(X)
```

```python
sns.scatterplot(df, x=df.sepal_width, y=df.petal_width,
hue=ocsvm.predict(X))
```
