```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
import plotly.express as px

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Run t-SNE
tsne = TSNE(n_components=3, random_state=7)
projections = tsne.fit_transform(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])

# Convert to DataFrame
proj_df = pd.DataFrame(projections, columns=['x', 'y', 'z'])
proj_df['species'] = df['species'].values

# Plot
fig = px.scatter_3d(proj_df, x='x', y='y', z='z', color='species', labels={'color': 'species'})
fig.show(renderer="browser")  # Use browser renderer to avoid notebook issues
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from umap import UMAP
import plotly.express as px

# Load Iris dataset
iris = load_iris()
features = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
species = pd.Categorical.from_codes(iris.target, iris.target_names)

# UMAP projections
umap_2d = UMAP(n_components=2, random_state=56)
umap_3d = UMAP(n_components=3, random_state=43)

proj_2d = umap_2d.fit_transform(features)
proj_3d = umap_3d.fit_transform(features)

# Convert to DataFrames
df_2d = pd.DataFrame(proj_2d, columns=['x', 'y'])
df_2d['species'] = species

df_3d = pd.DataFrame(proj_3d, columns=['x', 'y', 'z'])
df_3d['species'] = species

# Plot
fig_2d = px.scatter(df_2d, x='x', y='y', color='species')
fig_3d = px.scatter_3d(df_3d, x='x', y='y', z='z', color='species')

fig_2d.show(renderer="browser")  # Use browser renderer to avoid notebook issues
fig_3d.show(renderer="browser")
```
