```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV with correct delimiter and quoting
car = pd.read_csv("mpg.csv", quotechar='"', skipinitialspace=True)

# Normalize column names
car.columns = car.columns.str.strip().str.lower()

# Confirm column names
print(car.columns.tolist())  # 'cylinders' should be in this list
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define correct column names manually
columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
           'acceleration', 'model_year', 'origin', 'name']

# Load the CSV with space as delimiter and no header
car = pd.read_csv("mpg.csv", names=columns, header=None, delim_whitespace=True)

# Convert numeric columns to numeric (coerce errors to NaN)
numeric_cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']
for col in numeric_cols:
    car[col] = pd.to_numeric(car[col], errors='coerce')

# Drop rows with missing values in key columns
car = car.dropna(subset=['mpg', 'horsepower', 'cylinders'])

print("Data loaded successfully")
print(car.head())
```

```python
car.isnull().sum()
```

```python
sns.kdeplot(data=car, x='horsepower')
```

```python
sns.kdeplot(data=car, x='horsepower', bw_adjust=0.5)
```

```python
sns.kdeplot(data=car, x='mpg', y='horsepower')
```

```python
sns.kdeplot(data=car, x='mpg', y='horsepower', hue='cylinders')
```

```python
plt.figure(figsize=(20, 15))
sns.kdeplot(data=car, x='mpg', y='horsepower', hue='cylinders', fill=True, cbar=True)
```
